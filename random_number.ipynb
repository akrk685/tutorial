```python
# 데이터 샘플링
# 이미 있는 데이터 집합에서 일부를 무작위로 선택하는 것을 표본선택 혹은 샘플링
# choice 함수 사용
```

```python
 # 난수 생성
 # rand : 0부터 1사이의 균일 분포
 # randn : 표준 정규 분포
 # randint : 균일 분포의 정수 난수
```

```python
# rand 함수는 0부터 1사이에서 균일한 확률 분포로 실수 난수를 생성
# 숫자 인수는 생성할 난수의 크기
# 여러개의 인수를 넣으면 해당 크기를 가진 행렬을 생성
```

```python
np.random.rand(10)
```

```python
np.random.rand(3,5)
```

```python
# randn 함수는 기댓값이 0이고 표준편차가 1인
# 표준 정규 분포(standard normal distribution)를 따르는 난수를 생성
```

```python
np.random.randn(10)
```

```python
np.random.randn(3,5)
```

```python
np.random.randint(10, size = 10)
```

```python
np.random.randint(10,20,size=10)
```

```python
np.random.randint(10, 20, size=(3,5))
```
