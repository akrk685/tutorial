pandas 입문강의

```python
# pandas
# 데이터를 분석할 때 가장 많이 쓰이는 라이브러리
# 행과 열을 쉽게 처리할 수 있는 함수를 제공하는 도구
# numpy보다 유연하게 수치 연산이 가능
```

```python
# series
# index와 values로 이루어진 1차원 배열
# 모든 유형의 데이터를 보유할 수 있음
# 인덱스를 지정해 줄 수 있음
# 인덱스의 길이와 데이터의 길이는 같음
# [명시적 인덱스]와 [암묵적 인덱스]가 있음
# loc : 인덱스값을 기반으로 행 데이터를 읽는다
# iloc : 정수 인덱스를 기반으로 행 데이터를 읽는다
```

```python
import numpy as np
import pandas as pd
ex01=np.arange(0,50,10)
ex01
```

```python
# np.arange(start, stop, [step])
# step : 생략시 '1'로 간주. 실수도 가능
# array 형태의 자료형 반환

# range(start, stop, [step])
# step : 생략시 '1'로 간주. 정수만 가능
# list 형태의 자료형 반환
```

```python
ex02 = pd.Series(ex01, index=['a','b','c','d','e'])
print(ex02)
```

```python
ex02['c']
```

```python
ex02.loc['d'] # loc : 인덱스값을 기반으로 행 데이터를 읽는다
```

```python
ex02.iloc[2] #iloc : 정수 인덱스를 기반으로 행 데이터를 읽는다
```
