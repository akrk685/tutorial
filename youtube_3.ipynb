Matplotlib 기본활용

```python
# 그래프 기본 명령어

# 그래프그리기 모듈 호출 import matplotlib.pyplot as plt
# plot : '가상의 공간에 그래프를 그려라' 의미
# plt.show() : '완성된 그래프를 출력하라' 의미
# 그래프 사이즈 : plt.figure(figsize=(x,y)) # x,y값 지정
# x축제목 : plt.xlabel('name') # name값 지정
# y축제목 : plt.ylabel('name') # name값 지정
# 범례 : plt.legend()
# 그래프제목 : plt.title('name', fontsize = 30) # name값 지정, 30 : 글꼴크기
# 그래프출력하기 : plt.show()
```

```python
# 기본예제
```

```python
x = ['1반', '2반', '3반', '4반', '5반']
y1 = [84, 92, 83, 89, 90]
y2 = [84, 84, 91, 97, 70]
plt.figure(figsize=(10,5)) # 그래프 사이즈지정
plt.plot(x,y1,color='b', label='국어')
plt.bar(x,y2,color='g',label='영어')
plt.xlabel('학급')
plt.ylabel('점수')
plt.legend()
plt.title('과목별 점수 산출', fontsize=30)
plt.xticks(rotation=45)
plt.show()
```
