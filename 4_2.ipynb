```python
# CSV(Comma Separated Value) 
# 데이터 값이 쉼표로 구분되는 텍스트 파일
```

```python
import numpy as np
import pandas as pd
```

```python
%%writefile sample1.csv
c1, c2, c3
1, 1.11, one
2, 2.22, two
3, 3.33, three
```

```python
# CSV 파일로부터 데이터를 읽어 데이터프레임을 만들 때
pd.read_csv('sample1.csv')
# 0부터 시작하는 정수 인덱스가 자동으로 추가되었다.
```

```python
# 데이터 파일에 열 인덱스 정보가 없는 경우
%%writefile sample2.csv
1, 1.11, one
2, 2.22, two
3, 3.33, three
```

```python
pd.read_csv('sample2.csv', names=['c1', 'c2', 'c3'])
```

```python
pd.read_csv('sample1.csv')
```

```python
# 만약 테이블 내의 특정한 열을 행 인덱스로 지정
pd.read_csv('sample1.csv', index_col='c1')
```

```python
# 데이터를 구분하는 구분자(separator)가 쉼표가 아니면 sep 인수를 써서
# 구분자를 사용자가 지정

#길이가 정해지지 않은 공백이 구분자인 경우 -> /s+ 정규식 문자열을 사용
```

```python
%%writefile sample3.txt
c1         c2         c3         c4
0.179181  -1.538472   1.347553   0.43381
1.024209  0.087307    -1.281997  0.49265
0.417899  -2.002308   0.255245   -1.10515
```

```python
pd.read_table('sample3.txt', sep='\s+')
```

```python
# 자료 파일 중에 건너 뛰어야 할 행이 있으면 skiprows 인수를 사용
```

```python
%%writefile sample4.txt
파일 제목: sample4.txt
데이터 포맷의 설명:
c1, c2, c3
1, 1.11, one
2, 2.22, two
3, 3.33, three
```

```python
pd.read_csv('sample4.txt', skiprows=[0,1])
```

```python
# 특정한 값을 NaN으로 취급하고 싶으면 na_values 인수에 NaN 값으로 취급할 값을 넣는다
```

```python
%%writefile sample5.csv
c1, c2, c3
1, 1.11, one
2, , two
누락, 3.33, three
```

```python
df = pd.read_csv('sample5.csv', na_values=['누락'])
df
```

```python
# 파이썬의 데이터프레임 값을 CSV 파일로 출력하고 싶으면 to_csv 메서드를 사용
```

```python
df.to_csv('sample6.csv')
```

```python
df.to_csv('sample7.txt', sep='|')
```

```python
df.to_csv('sample8.csv', na_rep='누락')
```
